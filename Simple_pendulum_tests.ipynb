```python
import ODE_functions as DE
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Set parameters
## Define the physical parameters
# Define the mass of the bob
m = 0.2 # kg

# Define the length of the pendulum
l = 0.5 # m

# Define the gravitational constant
g = 9.81 # m/s^2

## Set simulation parameters
# Define the timestep
dt = 0.1
# Define the total time of the simulation
t_tot = 100 # sec

## Give initial values
th_init = 0.1 # rad
v_init = 0 # rad/s


```

```python
f = DE.f_simple_pend(m,l,g)

y_all, t_vect = DE.solveDE_RK4(f, y_init = [th_init, v_init], dt = dt, nr_of_DE = 2, t_tot = t_tot)
#y_all_Eu, t_vect = DE.solveDE_Euler(f, y_init = [th_init, v_init], dt = dt, nr_of_DE = 2, t_tot = t_tot)
```
